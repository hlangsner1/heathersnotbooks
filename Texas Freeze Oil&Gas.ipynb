```python
plt.figure(figsize=(10,10))

# Plot the returns
for stock in stocks: 
    plt.plot(returns[stock['ticker']], label=stock['name'])

# We need to call .legend() to show the legend.
plt.legend()
# Give the axes labels
plt.ylabel('Cumulative Returns %')
plt.xlabel('Time')
plt.show()
```
